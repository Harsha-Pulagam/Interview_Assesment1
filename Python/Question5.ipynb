```python
import requests
import pandas as pd

# API endpoint URL
url = 'http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes'

# Send GET request to the API endpoint
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the JSON data from the response
    data = response.json()

    # Access the necessary information from the data
    show_name = data['name']
    show_language = data['language']
    show_runtime = data['runtime']
    episode_data = data['_embedded']['episodes']
    data = pd.DataFrame(episode_data)
    data.head()

    ####print('Number of Episodes:', episode_count)
else:
    print('Error:', response.status_code)
```

```python
links = data["image"]
```

```python
links=pd.json_normalize(links)
```

```python
from bs4 import BeautifulSoup
def remove_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()
data["summary"]=data["summary"].apply(remove_html_tags)
```
