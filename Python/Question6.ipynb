```python
import pandas as pd
import numpy as np
df = pd.read_excel("pokemon.xlsx")
```

```python
#converting spawm time to time format
df['spawn_time'] = pd.to_datetime(df['spawn_time'], format='%M:%S').dt.time
```

```python
#extracting secs from spawm time
df['spam_time_sec'] = df['spawn_time'].apply(lambda x: x.minute * 60 + x.second)
```

```python
#calculating spwam rate
df['spam_rate'] = df[['spawn_chance', 'avg_spawns', 'spam_time_sec']].apply(lambda x: np.nan if x.isnull().any() else (x['spawn_chance'] * x['avg_spawns']*100)/x['spam_time_sec'], axis=1)
```

```python
#Pokemons whose spawn rate is less than 5%
df[df["spam_rate"] < 5]
```

```python
#counting the weakenesses
import ast
df['Weakness_count'] = df['weaknesses'].apply(lambda x: len(ast.literal_eval(x)))
```

```python
#Pokemons that have less than 4 weaknesses
df[df["Weakness_count"] < 4]
```

```python
#Pokemons that have no multipliers at all
df[df["multipliers"].isnull()]
```
